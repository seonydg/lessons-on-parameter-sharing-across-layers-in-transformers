- token preprocessing

```python
# token 중복 방지
word_list = set()
for data in dataset['validation']:
  for word in word_tokenize(data['translation']['en']):
    word_list.add(word)
  for word in word_tokenize(data['translation']['de']):
    word_list.add(word)
for data in dataset['test']:
  for word in word_tokenize(data['translation']['en']):
    word_list.add(word)
  for word in word_tokenize(data['translation']['de']):
    word_list.add(word)
```

```python
special_tokens = ['<SOS>', '<EOS>', '<PAD>', '<UNK>']

token2id = {}
id2token = set()

for token in word_list:
    id2token.add(token)
id2token = sorted(list(id2token))
if special_tokens:
    for token in special_tokens:
        id2token.insert(0, token)

for i, token in enumerate(id2token):
    token2id[token] = i
```

# dataset

- dataset

```python
from torch.utils.data import Dataset, DataLoader
```

```python
class NMTDataset(Dataset):
  def __init__(self, data, token2id):
    self.data = data
    self.src = []
    self.tgt = []
    self.src_mask = []
    self.tgt_mask = []

    max_len = 64

    for meta in data:
      temp1 = []
      mask=[]
      length=0

      # tokenization : pred
      for s in word_tokenize(meta['translation']['de'])[:62]: # sos, eos 추가 : 62 + 2 = 64(max length)
        length += 1
        try:
          temp1.append(token2id[s])
        except:
          temp1.append(token2id['<UNK>'])
      temp1.insert(0, token2id['<SOS>'])
      temp1.append(token2id['<EOS>'])
      length += 2
      for _ in range(length):
        mask.append(1)
      for _ in range(max_len - length):
        temp1.append(token2id['<PAD>'])
        mask.append(0)

      # mask(1 연산, 0 연산 x)
      self.src.append(temp1)
      self.src_mask.append(mask)
      temp2 = []
      length = 0

      # tokenizer : target
      for s in word_tokenize(meta['translation']['en'])[:62]:
        length += 1
        try:
          temp2.append(token2id[s])
        except:
          temp2.append(token2id['<UNK>'])
      temp2.insert(0, token2id['<SOS>'])
      temp2.append(token2id['<EOS>'])
      length += 2
      mask = []
      for _ in range(length):
        mask.append(1)
      for _ in range(max_len - length):
        temp2.append(token2id['<PAD>'])
        mask.append(0)
      self.tgt.append(temp2)
      self.tgt_mask.append(mask)


  def __len__(self):
    return len(self.data)


  def __getitem__(self, idx):
    return torch.tensor(self.src[idx]), torch.tensor(self.src_mask[idx]), torch.tensor(self.tgt[idx]), torch.tensor(self.tgt_mask[idx])
```

```python
train_data = NMTDataset(dataset['validation'], token2id)
valid_data = NMTDataset(dataset['test'], token2id)
```

- dataloader

```python
train_loader = DataLoader(train_data, shuffle=True, batch_size=64)
valid_loader = DataLoader(valid_data, shuffle=True, batch_size=64)
```

# Modeling

```python
from torch import nn


class ParameterSharedTransformerEncoder(nn.TransformerEncoder):
    def __init__(
        self,
        d_model=512,
        nhead=8,
        dim_feedforward=2048,
        dropout=0.1,
        activation="relu",
        num_unique_layers=3, # M
        num_total_layers=6, # N : 총 레이어 갯수
        mode="cycle_rev",
        norm=False,
    ):
        assert mode in {"sequence", "cycle", "cycle_rev"}
        quotient, remainder = divmod(num_total_layers, num_unique_layers)
        assert remainder == 0
        if mode == "cycle_rev":
            assert quotient == 2
        encoder_layer = nn.TransformerEncoderLayer(
            d_model, nhead, dim_feedforward, dropout, activation, batch_first=True
        )
        super().__init__(encoder_layer, num_layers=num_unique_layers, norm=norm)
        self.N = num_total_layers
        self.M = num_unique_layers
        self.mode = mode
        self.norm = nn.LayerNorm(d_model) if norm else None

    def forward(self, x, mask=None, src_key_padding_mask=None, verbose=False):
        for i in range(self.N):
            if self.mode == "sequence":
                i = i // (self.N // self.M)
            elif self.mode == "cycle":
                i = i % self.M
            elif i > (self.N - 1) / 2:
                i = self.N - i - 1
            if verbose:
                print(f"layer {i}")
            x = self.layers[i](x, mask, src_key_padding_mask)
        if self.norm is not None:
            x = self.norm(x)
        return x

```

- encoder / decoder

```python
encoder = ParameterSharedTransformerEncoder()

decoder_layer = nn.TransformerDecoderLayer(d_model=512, nhead=8, batch_first=True)
decoder = nn.TransformerDecoder(decoder_layer, num_layers=6)

classification_layer = nn.Linear(512, len(token2id))
```

- param

```python
de_params = decoder.parameters()
```

- embedding

```python
embedding = nn.Embedding(len(token2id), 512, padding_idx=1)
```

# Train

```python
epochs = 5
learning_rate=0.005
```

```python
optimizer = torch.optim.Adam(list(embedding.parameters())+list(encoder.parameters())+list(de_params)+list(classification_layer.parameters()), lr=learning_rate)
```

```python
loss_func = nn.CrossEntropyLoss(ignore_index=1)
```

- train

```python
for epoch in range(1, epochs):
  for i, batch_data in enumerate(train_loader):
    optimizer.zero_grad()
    src, src_mask, tgt, tgt_mask = batch_data
    emb = embedding(src)
    tgt_emb = embedding(tgt)
    memory = encoder(emb, src_mask.bool())
    out = decoder(tgt_emb, memory, tgt_mask.bool(), src_mask.bool())
    out = classification_layer(out.view(-1,512))
    loss = loss_func(out, tgt.view(-1))
    loss.backward()
    optimizer.step()
```

# evaluation

```python
from nltk.translate.bleu_score import sentence_bleu
```

```python
reference = [["this", "is", "the", "not", "sample", "banana"]]
candidate = ['this', "is", "the", "sample", "apple"]

score = sentence_bleu(reference, candidate, weights=(0.25, 0.25, 0.25, 0.25))
```

```python
for i, batch_data in enumerate(valid_loader):
  src, src_mask, tgt, tgt_mask = batch_data
  emb = embedding(src)
  enc_vector = encoder(emb, src_mask.bool())
  length = 1
  ys = torch.ones(src.size(0), 1).fill_(token2id['<SOS>']).type(torch.long)
  while length < 64: # max sequence length
      tgt_emb = embedding(ys)
      out = decoder(tgt_emb, enc_vector)
      prob = classification_layer(out[:, -1])
      next_word = torch.argmax(prob, dim=-1)
      next_word = next_word.item()
      ys = torch.cat([ys, torch.ones(1, 1).type(torch.long).fill_(next_word)], dim=1)

      len += 1
      if next_word == token2id['<EOS>']:
          break

  sentence = [id2token(w) for w in ys]
  target = [id2token(w) for w in tgt]
  score = sentence_bleu(sentence, target)
```
